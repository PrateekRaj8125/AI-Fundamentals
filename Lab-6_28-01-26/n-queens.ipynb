## N-Queens Problem

The n-queens puzzle is the problem of placing `n` queens on an `n x n` chessboard such that no two queens attack each other.

Given an integer `n`, return all distinct solutions to the ***n-queens puzzle***. You may return the answer in **any order**.

Each solution contains a distinct board configuration of the n-queens' placement, where `'Q'` and `'.'` both indicate a queen and an empty space, respectively.

 

Example 1:  
![Example 1](https://assets.leetcode.com/uploads/2020/11/13/queens.jpg)  
    **Input:** n = 4  
    **Output:** [[".Q..","...Q","Q...","..Q."],["..Q.","Q...","...Q",".Q.."]]  
    **Explanation:** There exist two distinct solutions to the 4-queens puzzle as shown above  

Example 2:  
    **Input:** n = 1
    **Output:** [["Q"]]


**Constraints:**  
- `1 <= n <= 9`

## Solution:

```python
import matplotlib.pyplot as plt
import numpy as np

class Solution:
    def solveNQueens(self, n: int):
        ans = []
        board = ["." * n for _ in range(n)]

        leftRow = [0] * n
        upperDiagonal = [0] * (2 * n - 1)
        lowerDiagonal = [0] * (2 * n - 1)

        self.solve(0, board, ans, leftRow, upperDiagonal, lowerDiagonal, n)
        return ans

    def solve(self, col, board, ans, leftRow, upperDiagonal, lowerDiagonal, n):
        if col == n:
            ans.append(board[:])  # deep copy
            return

        for row in range(n):
            if (leftRow[row] == 0 and
                lowerDiagonal[row + col] == 0 and
                upperDiagonal[n - 1 + col - row] == 0):

                board[row] = board[row][:col] + 'Q' + board[row][col + 1:]
                leftRow[row] = 1
                lowerDiagonal[row + col] = 1
                upperDiagonal[n - 1 + col - row] = 1

                self.solve(col + 1, board, ans, leftRow, upperDiagonal, lowerDiagonal, n)

                board[row] = board[row][:col] + '.' + board[row][col + 1:]
                leftRow[row] = 0
                lowerDiagonal[row + col] = 0
                upperDiagonal[n - 1 + col - row] = 0


def plot_chessboard(board):
    n = len(board)
    chessboard = np.zeros((n, n))

    for i in range(n):
        for j in range(n):
            if (i + j) % 2 == 0:
                chessboard[i][j] = 1

    plt.figure()
    plt.imshow(chessboard)
    plt.xticks(range(n))
    plt.yticks(range(n))
    plt.grid(True)

    # place queens
    for i in range(n):
        for j in range(n):
            if board[i][j] == 'Q':
                plt.text(j, i, '♛', ha='center', va='center', fontsize=24)

    plt.title("N-Queens Solution")
    plt.show()


def main():
    n = int(input("Enter n: ").strip())
    solution = Solution()
    results = solution.solveNQueens(n)

    for idx, board in enumerate(results, 1):
        print(f"Solution {idx}:")
        for row in board:
            print(row)
        print()

        plot_chessboard(board)


if __name__ == "__main__":
    main()
```
