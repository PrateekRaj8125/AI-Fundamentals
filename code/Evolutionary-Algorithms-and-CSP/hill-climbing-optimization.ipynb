## 🧩 Question:

Implement the **Hill Climbing algorithm** to solve an optimization or search problem.

- Use heuristic evaluation
- Improve solution iteratively
- Handle local optima

```python
import itertools

# Number of cities
n = int(input("Enter number of cities: "))

# City names
cities = []
for i in range(n):
    city = input(f"Enter name of city {i+1}: ")
    cities.append(city)

# Distance matrix
print("\nEnter distance matrix:")

dist = {}
for i in range(n):
    for j in range(n):
        d = int(input(f"Distance from {cities[i]} to {cities[j]}: "))
        dist[(cities[i], cities[j])] = d

# Cost calculation
def tour_cost(tour):
    cost = 0
    for i in range(len(tour)-1):
        cost += dist[(tour[i], tour[i+1])]
    return cost

# Generate tours starting from first city
start = cities[0]
tours = []

for perm in itertools.permutations(cities):
    if perm[0] == start:
        tours.append(list(perm))

best_tour = tours[0]
best_cost = tour_cost(best_tour)

for t in tours:
    cost = tour_cost(t)
    if cost < best_cost:
        best_cost = cost
        best_tour = t

print("\nBest Tour:", " -> ".join(best_tour))
print("Minimum Distance:", best_cost)
```
