## 🧩 Question:
Solve a **Cryptarithmetic problem** (e.g., SEND + MORE = MONEY) using appropriate AI techniques.

- Assign digits to letters
- Ensure arithmetic correctness
- Avoid duplicate digit assignments

```python
from itertools import permutations

def solve_cryptarithmetic(words, result):
    
    # Get unique characters
    unique_chars = set("".join(words) + result)
    unique_chars = list(unique_chars)
    
    if len(unique_chars) > 10:
        print("Too many unique characters")
        return
    
    # Leading characters cannot be zero
    leading_chars = set(word[0] for word in words + [result])
    
    digits = range(10)

    for perm in permutations(digits, len(unique_chars)):
        
        char_digit = dict(zip(unique_chars, perm))
        
        # Check leading zero constraint
        if any(char_digit[ch] == 0 for ch in leading_chars):
            continue
        
        # Convert words to numbers
        word_values = []
        for word in words:
            value = int("".join(str(char_digit[ch]) for ch in word))
            word_values.append(value)
        
        result_value = int("".join(str(char_digit[ch]) for ch in result))
        
        if sum(word_values) == result_value:
            print("Solution Found:")
            print(char_digit)
            print(f"{word_values[0]} + {word_values[1]} = {result_value}")
            return
    
    print("No solution found")


# Example 1
print("SEND + MORE = MONEY")
solve_cryptarithmetic(["SEND", "MORE"], "MONEY")

print("\nCROSS + ROADS = DANGER")
solve_cryptarithmetic(["CROSS", "ROADS"], "DANGER")
```
