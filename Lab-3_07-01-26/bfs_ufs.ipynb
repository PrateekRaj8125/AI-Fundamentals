## Uniform Cost Search for Optimal Path

**Objective:** Implement Uniform Cost Search for a weighted graph.

**Problem Statement:** Given a weighted graph (e.g., a transportation network with travel costs), find the minimum-cost path between two nodes.

**Tasks:**  
● Represent the graph as an adjacency list.  
● Implement Uniform Cost Search to find the optimal path.  
● Compare it with BFS for unweighted graphs.  

```python
import heapq
from collections import deque
# -------- Uniform Cost Search --------
def uniform_cost_search(graph, start, goal):
    frontier = [(0, start, [start])]  # (cost, node, path)
    visited_costs = {start: 0}
    while frontier:
        cost, node, path = heapq.heappop(frontier)
        if node == goal:
            return cost, path
        for neighbor, edge_cost in graph.get(node, []):
            new_cost = cost + edge_cost
            if neighbor not in visited_costs or new_cost < visited_costs[neighbor]:
                visited_costs[neighbor] = new_cost
                heapq.heappush(frontier, (new_cost, neighbor, path + [neighbor]))
    return float("inf"), []
```

```python
# -------- Breadth First Search --------
def bfs(graph, start, goal):
    queue = deque([(start, [start])])
    visited = {start}
    while queue:
        node, path = queue.popleft()
        if node == goal:
            return path
        for neighbor, _ in graph.get(node, []):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, path + [neighbor]))
    return []
```

```python
# -------- Graph Input Section --------
graph = {}
n = int(input("Enter number of nodes: "))
for _ in range(n):
    node = input("Enter node name: ")
    graph[node] = []
    edges = int(input(f"Enter number of edges from {node}: "))
    for _ in range(edges):
        neighbor, cost = input("Enter neighbor and cost: ").split()
        graph[node].append((neighbor, int(cost)))
start = input("Enter start node: ")
goal = input("Enter goal node: ")
```

```python
# -------- Algorithm Calls --------
cost, ucs_path = uniform_cost_search(graph, start, goal)
print("\nUniform Cost Search:")
print("Cost:", cost)
print("Path:", ucs_path)
bfs_path = bfs(graph, start, goal)
print("\nBreadth First Search:")
print("Path:", bfs_path)
```
