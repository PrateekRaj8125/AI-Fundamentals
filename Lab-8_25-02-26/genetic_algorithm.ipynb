```python
import random
import math

POP_SIZE = int(input("Enter population size: "))
GENS = int(input("Enter number of generations: "))
Pc = float(input("Enter crossover rate: "))
Pm = float(input("Enter mutation rate: "))

CHROM_LEN = 10

# Fitness function
def fitness(x):
    return math.sin(x) + x

# Decode binary to x
def decode(chrom):
    decimal = int(chrom,2)
    return (decimal/(2**CHROM_LEN-1))*10

# Initialize population
population = [''.join(random.choice('01') for _ in range(CHROM_LEN)) for _ in range(POP_SIZE)]

# Tournament selection
def tournament(pop):
    k = 3
    competitors = random.sample(pop,k)
    competitors.sort(key=lambda c: fitness(decode(c)), reverse=True)
    return competitors[0]

# One-point crossover
def crossover(p1,p2):
    if random.random() < Pc:
        point = random.randint(1,CHROM_LEN-1)
        c1 = p1[:point] + p2[point:]
        c2 = p2[:point] + p1[point:]
        return c1,c2
    return p1,p2

# Mutation
def mutate(chrom):
    new = ""
    for bit in chrom:
        if random.random() < Pm:
            new += '1' if bit=='0' else '0'
        else:
            new += bit
    return new

# GA loop
for g in range(GENS):

    new_pop = []

    while len(new_pop) < POP_SIZE:

        p1 = tournament(population)
        p2 = tournament(population)

        c1,c2 = crossover(p1,p2)

        c1 = mutate(c1)
        c2 = mutate(c2)

        new_pop.extend([c1,c2])

    population = new_pop[:POP_SIZE]

best = max(population, key=lambda c: fitness(decode(c)))

x_best = decode(best)

print("\nBest Chromosome:", best)
print("Best x:", round(x_best,4))
print("Maximum f(x):", round(fitness(x_best),4))
```
