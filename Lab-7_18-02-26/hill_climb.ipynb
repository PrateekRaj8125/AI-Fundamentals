## Hill Climb Search Algorithm

Implementation of local beam search algorithm using hill climb

```python
# Local Beam Search using Hill Climbing Concept
# Problem: Maximize the function f(x) = -(x-3)^2 + 10

import random
import matplotlib.pyplot as plt
import numpy as np

# Objective function
def objective_function(x):
    return -(x - 3)**2 + 10

# Generate neighbors (small random step)
def generate_neighbors(x, step_size=0.5):
    return [x + random.uniform(-step_size, step_size) for _ in range(5)]

# Local Beam Search Algorithm
def local_beam_search(beam_width, iterations):
    states = [random.uniform(-10, 10) for _ in range(beam_width)]
    best_progress = []

    for i in range(iterations):
        all_neighbors = []
        for state in states:
            neighbors = generate_neighbors(state)
            all_neighbors.extend(neighbors)

        # Select top beam_width states (Hill Climbing selection)
        states = sorted(all_neighbors, key=objective_function, reverse=True)[:beam_width]

        best_state = states[0]
        best_progress.append(objective_function(best_state))

    return states[0], objective_function(states[0]), best_progress


# -------- Example User Inputs --------
beam_width = int(input("Enter Beam Width:"))       # Example input: Beam Width
iterations = int(input("Enter No. of iterations:"))      # Example input: Number of iterations

print("User Inputs:")
print("Beam Width =", beam_width)
print("Iterations =", iterations)

best_x, best_value, progress = local_beam_search(beam_width, iterations)

print("\nBest Solution Found:")
print("x =", round(best_x, 4))
print("f(x) =", round(best_value, 4))

# -------- Plot Optimization Progress --------
plt.figure()
plt.plot(range(1, iterations + 1), progress)
plt.title("Local Beam Search Optimization Progress")
plt.xlabel("Iteration")
plt.ylabel("Best Function Value")
plt.show()
```
